```python
from scipy.special import zeta
from numpy import arange


test_vals = arange(1, 2, .01) 
```

```python
import matplotlib.pyplot as plt

outputs = zeta(test_vals)

plt.plot(test_vals, outputs)

plt.title("Zeta Function Input Plotted Against Output")

plt.xlabel("Input")

plt.ylabel("Zeta Output")

plt.savefig("Week_7/zeta_output.png")
```
